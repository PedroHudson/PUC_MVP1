```python
# Importacao do pandas
import pandas as pd

# Importacao do Numpy
import numpy as np
```

```python
# importando dados de uma url para um dataframe

# url a importar
url_dados = 'https://www.gov.br/receitafederal/dados/arrecadacao-estado.csv?raw=true'

# carga do dataset atraves da url
ipi = pd.read_csv(url_dados, sep=';',encoding='ISO-8859-1',decimal=',')

ipi_cols1=ipi[['Ano','Mês','UF','IPI - FUMO','IPI - BEBIDAS', 'IPI - AUTOMÓVEIS']]

ipi_cols1.dtypes

```

```python
ipi_cols1['IPI - FUMO']=ipi_cols1['IPI - FUMO'].str.replace(',','').astype(int)

ipi_cols1.describe()
```

```python
ipi_cols1['IPI - BEBIDAS']=ipi_cols1['IPI - BEBIDAS'].str.replace(',','').astype(int)

ipi_cols1.describe()
```

```python
ipi_cols1['IPI - AUTOMÓVEIS']=ipi_cols1['IPI - AUTOMÓVEIS'].str.replace(',','').astype(int)
```

```python
ipi_cols1.describe()
```

```python
ipi_cols1['Ano']=ipi_cols1['Ano'].astype(str)
```
